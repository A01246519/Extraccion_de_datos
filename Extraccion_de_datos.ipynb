```python
#Carga desde un archivo .xlxs  sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Filtro 1
filtro1=df[df["NOMBRE_VENDEDOR"]== "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
#IMPORTAR FILTRO 1

filtro1.to_csv("Actividad1_filtro1.csv")
```
