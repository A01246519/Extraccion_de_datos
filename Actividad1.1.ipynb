```python
#Filtro 8 - Y 
filtro8=df[(df["PRECIO_UNITARIO"] > 5000) & (df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN")]
filtro8
```

```python
#Filtro 9 - O 

filtro9=df[(df["PRECIO_UNITARIO"] > 5000) | (df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN")]
filtro9
```
